```python
import pandas as pd
import numpy as np
import os
import json
```

```python
carpeta_data = 'data/'
carpeta_players = 'players/'
carpeta_datos_formateados = 'Datos_Formateados/'
# Obtén todos los archivos en la carpeta
archivos = os.listdir(carpeta_data)

for archivo in archivos:
    datos_jugadores_por_equipo = []
    diccionario_de_jugadores = {}
    datos_equipo = pd.read_json(carpeta_data + archivo)
    for indice, fila in datos_equipo.iterrows():
        diccionario_de_jugadores[indice] = {'id': fila['id'], 'nombre': fila['slug']}
    for indice, jugador in diccionario_de_jugadores.items():
        nombre_equipo = ""
        with open(carpeta_players + archivo.split('.')[0] + '/' + str(jugador['id']) + '_' + jugador['nombre'] + '.json') as project_file:    
            datos_jugador = json.load(project_file)  
        nombre_equipo = datos_jugador['team']['name']
        del datos_jugador['marketValue'] #borrar estos datos que no nos sirven
        datos_jugadores_por_equipo.append(datos_jugador)
    
    json_equipo_data = {
    "nombre_equipo": nombre_equipo,
    "datos_jugadores": datos_jugadores_por_equipo
    }
    with open('Datos_Formateados/'+nombre_equipo+'.json', 'w') as json_file:
        json.dump(json_equipo_data, json_file, indent=4)
    
```
