```python
import pandas as pd

def dividir_por_temporadas(dataframe: pd.DataFrame, temporada):
    partidos_de_temporada = dataframe[dataframe["SEASON"] == temporada]
    return partidos_de_temporada
    
def get_jugadores(dataframe: pd.DataFrame):
    jugadores = dataframe["PLAYER_ID"].drop_duplicates()
    return jugadores

def get_estadisticas_jugadores(dataframe: pd.DataFrame):
    tiros_de_2_acertados = 0
    tiros_de_3_acertados = 0
    rebotes = 0
    asistencias = 0
    steals = 0
    for idx, partido in dataframe.iterrows():
        tiros_de_2_acertados += (partido['FGM'])
        tiros_de_3_acertados += (partido['FG3M'])
        rebotes += (partido['REB'])
        asistencias += (partido['AST'])
        steals += (partido['STL'])
        dataframe.at[idx, 'FGM_Total'] = tiros_de_2_acertados
        dataframe.at[idx, 'FG3M_Total'] = tiros_de_3_acertados
        dataframe.at[idx, 'REB_Total'] = rebotes
        dataframe.at[idx, 'AST_Total'] = asistencias
        dataframe.at[idx, 'STL_Total'] = steals
    return dataframe

dataframe_games = pd.read_csv("games.csv")
dataframe_details = pd.read_csv("games_details.csv")
dataframe_details.fillna(0, inplace=True)
temporada = 2003
while temporada < 2023:
    partidos_temporada = dividir_por_temporadas(dataframe=dataframe_games, temporada=temporada)
    partidos_temporada = partidos_temporada.iloc[::-1]
    partidos_temporada.reset_index(drop=True, inplace=True)
    details_partidos_temporada = dataframe_details[dataframe_details['GAME_ID'].isin(partidos_temporada['GAME_ID'])]
    jugadores = get_jugadores(dataframe=details_partidos_temporada)
    for jugador in jugadores:
        datos_jugador = details_partidos_temporada[details_partidos_temporada['PLAYER_ID'] == jugador]
        datos_jugador = datos_jugador.iloc[::-1]
        datos_jugador = get_estadisticas_jugadores(dataframe=datos_jugador)
        for idx, fila in datos_jugador.iterrows():
            fila_de_details_partidos = details_partidos_temporada.loc[(details_partidos_temporada['PLAYER_ID'] == fila['PLAYER_ID']) & 
                                                                      (details_partidos_temporada['TEAM_ID'] == fila['TEAM_ID']) & 
                                                                      (details_partidos_temporada['GAME_ID'] == fila['GAME_ID'])]
            indice = fila_de_details_partidos.index[0]
            details_partidos_temporada.at[indice, 'FGM_Total'] = fila['FGM_Total']
            details_partidos_temporada.at[indice, 'FG3M_Total'] = fila['FG3M_Total']
            details_partidos_temporada.at[indice, 'AST_Total'] = fila['AST_Total']
            details_partidos_temporada.at[indice, 'REB_Total'] = fila['REB_Total']
            details_partidos_temporada.at[indice, 'STL_Total'] = fila['STL_Total']
    details_partidos_temporada.to_csv('Datos_Formateados/'+str(temporada+1)+'-details.csv')
    temporada += 1
    

        
```
